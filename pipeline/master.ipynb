```python
import pandas as pd
import numpy as np
import glob
```

```python


# Get all Excel files ending with _ratios.xlsx in the current folder
files = glob.glob("*_ratios.xlsx")

dfs = [pd.read_excel(file) for file in files]
master_df = pd.concat(dfs, ignore_index=True)
master_df.to_excel("master_ratios.xlsx", index=False)


```

```python
# Get all Excel files ending with _inputs.xlsx in the current folder
input_files = glob.glob("*_inputs.xlsx")

# Read each file into a DataFrame
input_dfs = [pd.read_excel(file) for file in input_files]

# Concatenate them into one master dataframe
master_inputs_df = pd.concat(input_dfs, ignore_index=True)

master_inputs_df.to_excel("master_inputs.xlsx", index=False)
```
